```python
#importing libraries
import pandas as pd               #for analyse & manipulation of data.
import seaborn as sns             #Both matplotlib and seaborn are used for visualizations.
import matplotlib.pyplot as plt  
import numpy as np               #For arraya
from scipy.optimize import curve_fit   
import warnings
warnings.filterwarnings('ignore')
```

```python
csv_data_path = "C:/Users/CODERS READY/globalterrorism.csv"  #Location path of the file.
```

```python
globalterrorism = pd.read_csv(csv_data_path)  #reading the data file means import the csv file.
```

```python
rm = pd.DataFrame(globalterrorism)        #creating data frame.
rm
```

```python
print("There are {} rows and {} columns in the dataset.".format(rm.shape[0],rm.shape[1])) #showing total no. of rows and columns
```

Renaming some columns.

```python
rm = rm.rename(columns={'eventid':'Event id','iyear':'Year','imonth':'Month','iday':'Day','approxdate':'Approx date','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','attacktype1_txt':'Attack_type','targtype1_txt':'Target_type','targsubtype1_txt':'Target_subtype','gname':'Gang_name','weaptype1_txt':'Weapon_type'})
rm
```

### <font color='Violet'>NEW DATAFRAME(HERE FILTERATE DATA ACC. TO REQ.)</font>

```python
rm = rm[['Year','Month','Day','Country','Region','State','City','Attack_type','Target_type','Target_subtype','Gang_name','Weapon_type']]
rm
```

# <font color='Orange'>VISUALIZATIONS</font>
<B> some hot zones using visualizations. For this, we have to use matplotlib and seaborn which we have installed at the beginning.

```python
plt.figure(figsize=(22,6));
sns.countplot(x=rm.Weapon_type);
```

<B> Most of the attacks were attacked through explosives and then through firearms.

## <font color='blue'>ATTACKS BY YEAR</font>

```python
plt.figure(figsize=(20,5));
plt.xticks(rotation =30)
plt.yticks(rotation =60)
plt.title("Attacks By Year")
sns.countplot(x=rm.Year);
```

```python
pd.crosstab(rm.Year,rm.Region).plot(kind='area',figsize=(15,6))
plt.title('Terrorist Activities by Region in each Year')
plt.ylabel('Number of Attacks')
plt.show()
```

<b>In these graph we can see the terrorist activities by each region . so here Australasia and oceania in 2010 it has been in the highest number of terrorist attack.

## <font color='magenta'>ATTACKS BY MONTH</font>

```python
plt.figure(figsize=(15,6));
plt.title('ATTACKS BY MONTH',backgroundcolor='yellow')
plt.xticks(rotation =30,)
plt.yticks(rotation =30,)
sns.countplot(rm['Month'],palette= sns.color_palette("flare"));
```

## <font color='green'>ATTACKS BY DAY

```python
plt.figure(figsize=(20,6));
plt.title('ATTACKS BY DAY',backgroundcolor='yellow')
sns.countplot(x=rm.Day,palette = "mako");
```

## <font color='red'>ATTACKS BY COUNTRY

```python
plt.figure(figsize=(35,6));
plt.rcParams.update({'font.size':8});
plt.xticks(rotation = 90)
plt.title('Attacks by Country')
sns.countplot(x=rm.Country);
```

<B>Iraq dominates all the countries and it has the highest number of attacks and then Pakistan, Afghanistan, and India follow it. Government should increase safety measures in these countries.

## <font color='brown'>TARGET TYPES

```python
plt.figure(figsize=(15,7));
plt.rcParams.update({'font.size':10});
plt.title('Based on Target type')
sns.countplot(y=rm.Target_type,palette='inferno');
```

For most of the attacks, the target is Private Citizens& property and the next Military leads.

# WEAPON TYPES

```python
plt.figure(figsize=(15,6));
sns.kdeplot(x='Year' , data=rm, hue='Weapon_type')
```

## <font color='dark'>ATTACK TYPES

```python
atck=[]
[atck.append(x) for x in rm['Attack_type'] if x not in atck]
fig = plt.figure(figsize =(10,18))
explode = (0.1, 0.2, 0.1,0.2,0.2,0.4,0.5,0.7,0.9)
plt.pie(rm['Attack_type'].value_counts(),labels=atck,autopct='%1.2f%%',explode=explode)
plt.legend(title ="ATTACK TYPE",title_fontsize=10,loc ="best",fontsize=10,bbox_to_anchor =(1, 0, 0.5, 1))
plt.show()
```
